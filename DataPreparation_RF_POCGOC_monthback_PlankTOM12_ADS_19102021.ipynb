```python
def MLD_averages_1(df, df_inMLD_test):
    df_inMLD = df.loc[df['depth'] <= df['MLD']]
    df_inMLD_test = df_inMLD.groupby(['coordinates','year','month']).mean().reset_index()

    df_inMLD_test.rename(columns={"Temp": "Temp_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"Temp_back": "Temp_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"O2": "O2_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"O2_back": "O2_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"NO3": "NO3_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"NO3_back": "NO3_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"PO4": "PO4_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"PO4_back": "PO4_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"Si": "Si_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"Si_back": "Si_back_mean"}, inplace=True)    
    df_inMLD_test.rename(columns={"CHL": "CHL_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"CHL_back": "CHL_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"BAC": "BAC_mean"}, inplace=True)   
    df_inMLD_test.rename(columns={"BAC_back": "BAC_back_mean"}, inplace=True) 
    df_inMLD_test.rename(columns={"MES": "MES_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"MES_back": "MES_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"PTE": "PTE_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"PTE_back": "PTE_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"DIA": "DIA_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"DIA_back": "DIA_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"COC": "COC_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"COC_back": "COC_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"PIC": "PIC_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"PIC_back": "PIC_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"PHA": "PHA_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"PHA_back": "PHA_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"GEL": "GEL_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"GEL_back": "GEL_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"PRO": "PRO_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"PRO_back": "PRO_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"MAC": "MAC_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"MAC_back": "MAC_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"MIX": "MIX_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"MIX_back": "MIX_back_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"FIX": "FIX_mean"}, inplace=True)
    df_inMLD_test.rename(columns={"FIX_back": "FIX_back_mean"}, inplace=True)
    
    return df_inMLD_test
```
